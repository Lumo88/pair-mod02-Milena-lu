```python
import requests
import pandas as pd
```

```python
#url de la API
url_git = "https://api.github.com/users/gootyfer"
```

```python
#aqui hacemos la llamada
response = requests.get(url_git)
```

```python
#conocemos el estado de la conexión, 200 es éxito
response.status_code
```

```python
#vemos el contenido de esa llamad
# 
response.content
```

```python
#aqui printeamos en formato json
response.json()
```

```python
datos_git = response.json() #guardamos en variable el resultado
```

```python
len(datos_git) #un diccionario con 33 keys
```

```python
datos_git.keys() #mostramos todas las keys
```

```python
#login, name, location, followers, following

print(f"Usuario : {datos_git["login"]}")
print("Nombre: ",datos_git["name"])
print("Localización: ",datos_git["location"])
print("Followers: ",datos_git["followers"]," usuarios")
print("Following: ",datos_git["following"]," usuarios")
```

```python
type(datos_git)
```

```python
for key, value in datos_git.items():
    print (key,"---->",value)
```
